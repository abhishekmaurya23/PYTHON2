Sure! Here's a Python program that accepts a percentage from the user and displays the corresponding grade based on the given criteria:

```python
percentage = float(input("Enter the percentage: "))

if percentage > 90:
    grade = "A"
elif percentage > 80 and percentage <= 90:
    grade = "B"
elif percentage >= 60 and percentage <= 80:
    grade = "C"
else:
    grade = "D"

print("Grade:", grade)
```

In this program, the user is prompted to enter a percentage value using the `input()` function, which is then converted to a floating-point number using `float()`.

The program then uses a series of `if-elif-else` statements to determine the grade based on the given criteria. If the percentage is greater than 90, the grade is set to "A". If it falls between 80 and 90 (inclusive), the grade is set to "B". If it falls between 60 and 80 (inclusive), the grade is set to "C". Otherwise, if the percentage is below 60, the grade is set to "D".

Finally, the program displays the grade using the `print()` function.

Note: This program assumes that the user will enter a valid percentage value. Error handling for invalid inputs is not included in this example.

```python
city = input("Enter the name of the city: ")

monuments = {
    "Delhi": ["Red Fort", "Taj Mahal"],
    "Agra": ["Taj Mahal"],
    "Jaipur": ["Jal Mahal"]
}

if city in monuments:
    print("Monuments in", city + ":")
    for monument in monuments[city]:
        print(monument)
else:
    print("No monuments found for", city)
```

In this program, the user is prompted to enter the name of a city using the `input()` function, which is stored in the `city` variable.

The program then uses a dictionary `monuments` to map cities to their corresponding monuments. Each city is associated with a list of monuments.

The program checks if the entered city exists in the `monuments` dictionary. If it does, it iterates through the list of monuments for that city and displays them using a `for` loop. If the city is not found in the dictionary, it displays a message indicating no monuments were found.

For example, if the user enters "Delhi," the program will display "Red Fort" and "Taj Mahal" as the monuments associated with Delhi.

Sure! Here's a Python program that calculates how many times a given number can be divided by 3 before it becomes less than or equal to 10:

```python
number = int(input("Enter a number: "))
count = 0

while number > 10:
    number /= 3
    count += 1

print("The number can be divided by 3", count, "times before it is less than or equal to 10.")
```

In this program, the user is prompted to enter a number using the `input()` function, which is then converted to an integer using `int()` and stored in the `number` variable.

The program uses a `while` loop to repeatedly divide the number by 3 until it becomes less than or equal to 10. Each time the division occurs, the count is incremented by 1.

Finally, the program displays the count, indicating how many times the given number can be divided by 3 before it becomes less than or equal to 10.

For example, if the user enters 81, the program will output "The number can be divided by 3 4 times before it is less than or equal to 10."

The `while` loop in Python is used to execute a block of code repeatedly as long as a specified condition is true. It allows you to create loops where the number of iterations is not known beforehand or where the loop should continue until a certain condition is met. The `while` loop is useful when you want to repeat a block of code until a specific condition becomes false.

The general syntax of a `while` loop in Python is as follows:

```python
while condition:
    # Code block
    # (executed as long as the condition is true)
```

The `condition` is an expression that is evaluated before each iteration. If the condition is true, the code block inside the loop is executed. After the code block is executed, the condition is evaluated again. If the condition remains true, the loop continues to execute, and this process repeats until the condition becomes false.

Here's an example that demonstrates the usage of a `while` loop:

```python
count = 0

while count < 5:
    print("Count:", count)
    count += 1

print("Loop finished")
```

In this example, the `while` loop executes as long as the `count` variable is less than 5. Inside the loop, the current value of `count` is printed, and then `count` is incremented by 1 using `count += 1`. This process repeats until `count` becomes 5.

The output of the above code will be:
```
Count: 0
Count: 1
Count: 2
Count: 3
Count: 4
Loop finished
```

The `while` loop is especially useful in scenarios where the number of iterations is not predetermined, or when you want to repeatedly execute a block of code until a certain condition is met. However, it is important to ensure that the condition inside the `while` loop will eventually become false; otherwise, the loop will continue indefinitely, resulting in an infinite loop.

ANS:7-To reverse a `while` loop and display numbers from 10 to 1, you can modify the loop condition and the iteration step. Here's an example:

```python
count = 10

while count >= 1:
    print("Count:", count)
    count -= 1

print("Loop finished")
```

In this modified `while` loop, the `count` variable is initially set to 10. The loop condition `count >= 1` checks if the `count` is greater than or equal to 1. Inside the loop, the current value of `count` is printed, and then `count` is decremented by 1 using `count -= 1`. This process repeats until `count` becomes 1.

The output of the above code will be:
```
Count: 10
Count: 9
Count: 8
Count: 7
Count: 6
Count: 5
Count: 4
Count: 3
Count: 2
Count: 1
Loop finished
```

As a result, the loop starts from 10 and decrements `count` by 1 in each iteration, printing the numbers in reverse order until it reaches 1.
